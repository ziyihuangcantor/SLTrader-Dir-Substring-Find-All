```python
import os
import pandas as pd
```

```python
# Finds all .cshtml files that containing the target string

path = 'C:/Projects/StockLoans/Src/SLTrader/Areas'

target_str = 'html.text'.lower()

target_files = []

for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith('.cshtml')):
            with open(os.path.join(root, file), 'r') as f:
                content = f.read()
                if target_str in str(content).lower():
                    # Gets the full path
                    my_dir = str(os.path.join(root,file))
                    
                    # Cleans up the path
                    my_dir_trunc = my_dir.replace('C:/Projects/StockLoans/Src/SLTrader/', '~')
                    target_dir = my_dir_trunc.replace('\\', '/')
                    
                    print(target_dir)
                    
                    inner_dict = {}
                    inner_dict[target_dir] = []
                    
                    # Finds the target line numbers
                    with open(os.path.join(root, file), 'r') as f2:
                        lines = f2.readlines()
                        count = 0
                        for line in lines:
                            count = count + 1
                            if target_str in str(line).lower():
                                inner_dict[target_dir].append(str(count))
                                
                    target_files.append(inner_dict)
```

```python
# Seperates keys and values into two list

files_lst = []
numbers_lst = []

for element in target_files:
    curr_key = ''.join(list(element.keys()))
    curr_value = list(element.values())[0]
    files_lst.append(curr_key)
    numbers_lst.append(curr_value)
```

```python
import os
```

```python
# Finds all .cshtml files that containing the target string

path = 'C:/Users/ZH128895/Work/SLTraderDirSubstringFindAll/Areas'

target_str = '@html.text'.lower()

target_files = []

for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith('.cshtml')):
            with open(os.path.join(root, file), 'r') as f:
                content = f.read()
                if target_str in str(content).lower():
                    # Gets the full path
                    my_dir = str(os.path.join(root,file))
                    
                    # Clean up the path
                    my_dir_trunc = my_dir.replace('C:/Users/ZH128895/Work/SLTraderDirSubstringFindAll/', '~')
                    target_dir = my_dir_trunc.replace('\\', '/')
                    
                    inner_dict = {}
                    inner_dict[target_dir] = []
                    
                    
                    with open(os.path.join(root, file), 'r') as f2:
                        lines = f2.readlines()
                        count = 0
                        for line in lines:
                            count = count + 1
                            if target_str in str(line).lower():
                                inner_dict[target_dir].append(str(count))
                                
                    target_files.append(inner_dict)
                
```
